# One Hot Encoding with Loop

```python
for col in df8.columns:
    if is_numeric_dtype(df8[col]):                                   #If columns of df8 are numeric, then continue
        continue
    one = pd.get_dummies(df8[col], drop_first = True, prefix = 'Area')
    df8.drop(df8[[col]], axis = 1, inplace = True)
    df8 = pd.concat([df8, one], axis = 1)
    
   

    
# If the columns are not numeric, then it will generate dummy variable from that specific column and drop the first column.
# After getting dummy variable, we do not need the previous 'Area' column, so we can drop that column of df8 from df8.
# Then we will concat/add the df8 without Area column with the new dummy variable column.
```
